```python
from tkinter import *
import requests
import bs4
```

```python
root =Tk()

root.geometry('400x400')

#list of tuple to creat radio buttons

city_lst=[
    ("New Delhi", 1),
    ("Mumbai", 2),
    ("Bangalore",3),
    ("Kolkatta",4)
]

# creating lable to display our result

r_label=Label(root, text='', font=('Helvetica', 20), padx=10, pady=20)
r_label.pack()

#defining variable type we are going to use in Radiobutton
r =IntVar()

#defing function to get our result after selection of city
def r_value(value):
    
    city_index= value
    
    #deciding which websites should we have to scrape to get desire result
    if city_index == 1:
        api_req= requests.get("https://www.aqi.in/dashboard/india/delhi/new-delhi")
        
    elif city_index == 2:
        api_req= requests.get("https://www.aqi.in/dashboard/india/maharashtra/mumbai")
        
    elif city_index == 3:
        api_req= requests.get("https://www.aqi.in/dashboard/india/karnataka/bangalore")

    elif city_index == 4:
        api_req= requests.get("https://www.aqi.in/dashboard/india/west-bengal/kolkata")

    #By using Bautifule Soup Library to get necessary details...
    soup= bs4.BeautifulSoup(api_req.text, 'lxml')
    result= soup.select('.sectot-box-left')

    #getting the state name...
    state_txt= result[0].select('h4')[0].text
    state_name= state_txt.split(',')[0]

    #getting the AQI number for New Delhi State....
    aqi_number= result[0].select('h2')[0].text


    #getting the category...
    category= result[0].select('h3')[0].text.upper()

    # let set colour of background of window according to category of AQI
    if category== 'GOOD':
        weather_bg= 'green'
    elif category == 'MODERATE':
        weather_bg= 'Yellow'
    elif category == 'POOR':
        weather_bg= 'RED'
    elif category == 'UNHEALTHY':
        weather_bg= 'violet'
    elif category == 'SEVERE':
        weather_bg= 'Purple'
    elif category == 'HAZARDOUS':
        weather_bg= 'brown'
    else:
        pass


    r_label['text']= 'Air Quality Index Of '+ state_name + ' is '+ aqi_number+' and it is in '+category+' category.'
    r_label['bg']= weather_bg
 
# Creating Radio Button
for city, index in city_lst:
    r_btn= Radiobutton(root, text= city, variable=r, value=index)
    r_btn.pack(anchor=W)

    
btn= Button(root, text="Enter", width=15, command= lambda: r_value(r.get()))
btn.pack()

r_label= Label(root, text=r.get())
r_label.pack()
root.mainloop()
```
