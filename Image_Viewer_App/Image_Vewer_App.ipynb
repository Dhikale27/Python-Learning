```python
from tkinter import *
from PIL import ImageTk, Image
```

```python
root= Tk()

root.geometry('500x500')

img_lst =['img1.jpeg', 'img2.jpeg', 'img3.jpeg', 'img4.jpeg', 'img5.jpeg']

img= ImageTk.PhotoImage(Image.open('img1.jpeg'))

label_img= Label(root, image=img)
label_img.pack()

i=1
def forward():

    #global label_img
    
    #label_img.grid_forget()
    
    global i
    

    #btn_next.configure(state=NORMAL)


    img_add= img_lst[i]

    img= ImageTk.PhotoImage(Image.open(img_add))

    label_img.configure(image= img)
    label_img.image= img
    
    btn_back.configure(state=NORMAL)
    
    if i == len(img_lst)-1:
        btn_next.configure(state=DISABLED)

    i+=1
        
    
def backward():

    #global label_img
    
    #label_img.grid_forget()
    
    global i
    
    
    if i > len(img_lst)-1:
        i= i-1

    img_add= img_lst[i]

    img= ImageTk.PhotoImage(Image.open(img_add))

    label_img.configure(image= img)
    label_img.image= img
    i-=1
    
    btn_next.configure(state=NORMAL)
    
    if i < 0:
        btn_back.configure(state=DISABLED)
    
    #label_img.grid(row=0, column=0, columnspan=3)

    
    
btn_back= Button(root, text='<<', width=10, command= backward, bg='blue')
btn_back.pack(side= LEFT)

btn_next= Button(root, text='>>', width= 10, command=forward, bg='blue')
btn_next.pack(side= RIGHT)



root.mainloop()
```
